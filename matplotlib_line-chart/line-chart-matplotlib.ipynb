# Line Chart Using Python's Matplotlib

```python
from matplotlib import pyplot as plt
# NOTE: The pyplot module is a simple interface for matplotlib.
```

```python
plt.plot([1, 2, 3], [1, 4, 9])
plt.plot([1, 2, 3], [10, 20, 30])
plt.legend(['Data Set 1', 'Data Set 2'])
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Line Chart Using Matplotlib')
# plt.show()
plt.savefig('exported_image')
# this saves the file as a png
```
